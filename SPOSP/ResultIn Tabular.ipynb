### Learning Curve

```python
N,noise,deg,lr = 1000,0.5,4,0.5
twostagelr = pd.read_csv("LearningCurve/Normed_Twostage_data_N_{}_noise_{}_deg_{}_lr{}.csv".format(N,noise,deg,lr))
plt.figure(figsize=(10,6))
sns.lineplot(data= twostagelr, x="step", y="val_regret", hue="model")
plt.xlabel('Steps', fontsize = 18)
plt.ylabel('Validation Regret(%)', fontsize = 18)
plt.xticks(fontsize=18)
# plt.xlim(0, 750)
plt.legend( prop={'size': 16})
```

```python
N,noise,deg,lr = 1000,0.5,4,0.1
spolr = pd.read_csv("LearningCurve/Normed_SPO_data_N_{}_noise_{}_deg_{}_lr{}.csv".format(N,noise,deg,lr))
plt.figure(figsize=(10,6))
sns.lineplot(data= spolr, x="step", y="val_regret", hue="model")
plt.xlabel('Steps', fontsize = 18)
plt.ylabel('Validation Regret(%)', fontsize = 18)
plt.xticks(fontsize=18)
# plt.xlim(0, 750)
plt.legend( prop={'size': 16})
```
